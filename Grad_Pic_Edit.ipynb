```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# Import grad pic and watermark img
img = cv.cvtColor(cv.imread("imgAssets/grad_pic.png"), cv.COLOR_BGR2RGB)
watermark = cv.imread("imgAssets/watermark.png", 0)

# Get all contours of watermark img
ret, thresh = cv.threshold(watermark, 127, 255, 0)
im, contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

# Create mask using contours
mask = np.zeros((watermark.shape[0], watermark.shape[1]), dtype=np.uint8)
cv.drawContours(mask, contours, -1, (255,255,255), 2)

# Remove watermark
processedImg = cv.inpaint(img, mask, 3, cv.INPAINT_TELEA)

            
# Display original image
plt.imshow(img)
plt.title("Original Grad Pic")
plt.xticks([]), plt.yticks([])
plt.show()

# Display watermark image
plt.imshow(mask, cmap="gray")
plt.title("Watermark")
plt.xticks([]), plt.yticks([])
plt.show()

# Display edited grad pic
plt.imshow(processedImg)
plt.title("Edited Grad Pic")
plt.xticks([]), plt.yticks([])
plt.show()
```
