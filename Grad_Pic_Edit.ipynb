```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# Import grad pic and watermark img
img = cv.cvtColor(cv.imread("imgAssets/grad_pic.png"), cv.COLOR_BGR2RGB)
imgCopy = img.copy()
watermark = cv.imread("imgAssets/watermark.png", 0)

# Get all contours of watermark img
ret, thresh = cv.threshold(watermark, 127, 255, 0)
im, contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
cv.drawContours(imgCopy, contours, -1, (0,255,0), 2)

# Display original image
plt.imshow(img)
plt.title("Original Grad Pic")
plt.xticks([]), plt.yticks([])
plt.show()

# Display watermark image
plt.imshow(watermark, cmap="gray")
plt.title("Watermark")
plt.xticks([]), plt.yticks([])
plt.show()

# Display original image with watermark highlighted
plt.imshow(imgCopy)
plt.title("Highlighted Watermark")
plt.xticks([]), plt.yticks([])
plt.show()
```
